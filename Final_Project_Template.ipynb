```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn import preprocessing
from google.colab import files
```

## Define your clusters

```
# Values must be between 0 and 1.  
# For this example I have 3 prototypes in a 4 dimensional input space.  
# For your project you must have 5 prototype vectors in a 6 or more dimensional 
# input space. 
#
# Dim 1: Height
# Dim 2: Weight
# Dim 3: Friendliness
# Dim 4: Slobberiness
#
# Standard deviation is up to you; it will determine the success of your classifier.
# Smaller is easier for the algorithmm to deal with, but less realistic. You are 
# welcome to use categories that overlap slightly.
# 
C = [[.1, .1, .2,.1], [.5, .5, .9,.4], [.8,.8,.6,.9]]
cluster_labels = ["chihuaha", "golden retriever", "mastiff"]
inputs, targets = make_blobs(n_samples=1000, centers=C, 
                  cluster_std=.05, 
                  random_state=2)
```

## Plot a few salient dimensions

```
plt.figure(figsize=(7, 7))
# Define your dimensions below
height_dim = 1 
weight_dim = 2
aggression_dim = 3
slobber_dim = 4
# Decide which two dimensions to plot below
x = aggression_dim - 1 
y = slobber_dim - 1
plt.scatter(inputs[:,x],inputs[:,y])
plt.xlabel("Friendliness") # Be sure to change these labels
plt.ylabel("Slobberiness")
for i in range(len(C)):
    plt.annotate(cluster_labels[i], (C[i][x], C[i][y]))
```

# Project using dimensionality reduction

```
# The dimensionality reduction is PCA, which is also what the Simbrain projection plot uses by default
proj = PCA(n_components=2).fit_transform(inputs)
proj_centers = PCA(n_components=2).fit_transform(C)
```

```
plt.scatter(proj[:,0],proj[:,1]);
for i in range(len(C)):
    plt.annotate(cluster_labels[i], (proj_centers[i][0], proj_centers[i][1]))
```
